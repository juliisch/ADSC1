```python
df = df.rename(columns={"nbhd": "nbhd_1",
                        "taxkey": "taxkey_1"})
```

```python
# Spaltenbezeichnung in Kleinbuchstaben umwandeln
df.columns = [x.lower() for x in df.columns.to_list()]
```

```python
# Duplikate l√∂schen
df = df.drop_duplicates()
```

```python
# Subset von proptype auf "Residential" (Wohnhaus)
df = df[df["proptype"] == "Residential"]
df = df.drop(columns=['proptype'])
```

```python
# Anzahl der NaN-Werte pro Spalte (Absolut und Relativ)

# Anzhal NaN 
count_nan = df.isna().sum()
# Prozentsatz NaN
nan_percent = round((count_nan / len(df)) * 100,2)
# DataFrame
df_nan = pd.DataFrame({
    'Anzahl NaN': count_nan,
    'Prozent NaN_': nan_percent
}).sort_values(by='Anzahl NaN', ascending=False)
list_var_toDelete_NaN = df_nan[df_nan["Prozent NaN_"] >= 40].index.tolist()
df_nan[df_nan["Prozent NaN_"] >= 40] # Ausgeben die √ºber 40% 
```

```python
# Nicht ben√∂tigte Variablen bzw. Variablen mit zu vielen NaN-Values l√∂schen
df = df.drop(columns=list_var_toDelete_NaN)
```

```python
# Unrealistische Werte auf NaN setzten
df.loc[df['year_built'] < 1000, 'year_built'] = np.nan
df.loc[df['bdrms'] > 100, 'bdrms'] = np.nan
```

```python
unlabeled = df[df["sale_price"].isna()].copy()
print(f"Der Datensatz hat {len(unlabeled)} ungelabeldete Eintr√§ge")
labeled = df[df["sale_price"].notna()]
print(f"Der Datensatz hat {len(labeled)} gelabeldete Eintr√§ge")

# NaN-Werte der Spalze Sale_price werden gel√∂scht, da es sich hierbei um die Zielvariable hadnelt
df = df.dropna(subset=['sale_price'])
```

```python
# NaN Werte der Kategorischen Variablen auf 'Unknown' setzten
#df['proptype'] = df['proptype'].fillna('Unknown')
df['style'] = df['style'].fillna('Unknown')
df['extwall'] = df['extwall'].fillna('Unknown')
df['address'] = df['address'].fillna('Unknown')
```

```python
cols_with_nan = df.columns[df.isna().any()].tolist()
df[cols_with_nan].isna().sum()
```

```python
 # Spalten die eig. ganzzahlig sind, werden auf das Datenformat Int ge√§ndert
for col in df.select_dtypes(include="float"):
    if (df[col].dropna() % 1 == 0).all():  # Pr√ºfung, ob alle Werte auf .0 enden (ganzzahlig sind), dabei werden NaN Werte nicht betrachtet
        df[col] = df[col].astype("Int64")

# Datentyp von Lotsize in Int √§ndern
df['lotsize'] = pd.to_numeric(df['lotsize'].astype(str).str.replace(',', '', regex=False),errors='coerce').astype('Int64')
# Datentyp von Sale_price in Int √§ndern
df['sale_price'] = pd.to_numeric(df['sale_price'].astype(str).str.replace('[^\d]', '', regex=True), errors='coerce').astype('Int64')
df.loc[df['sale_price'] >= 408770000, 'sale_price'] = np.nan
df = df.dropna(subset=['sale_price'])

# Datentyp von sale_date auf Datetime √§ndern
#df["sale_date"] = pd.to_datetime(df["sale_date"], format="%Y-%m", errors='coerce')
#df["sale_date"] = df["sale_date"].dt.month
df = df.drop(columns=['sale_date'])

```

```python
# Bei den Adressen die Hausnummer entfernen
#df["address_nn"] = df["address"].str.split(" ", n=1).str[1]
#df = df.drop(columns=['address'])

df = df.drop(columns=['address'])
```

```python

df["extwall"] = df["extwall"].str.replace("-", "_").str.replace("/", "_").str.replace(" ", "")
```
