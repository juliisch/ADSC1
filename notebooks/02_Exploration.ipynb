Duplikate löschen

```python
# Duplikate löschen
df = df.drop_duplicates()
```

Filterung der Variable PropType auf "Residential" (Wohnhaus)

```python
# Selektierung von proptype auf "Residential" (Wohnhaus)
df = df[df["PropType"] == "Residential"]
df = df.drop(columns=['PropType'])
```

Variablen ohne Informationen in Bezug auf die Zeilvariable werden gelöscht

```python
#TODO Entscheidne was mit Address passier
list_var_toDelete = ['PropertyID', 'CondoProject', 'Sale_date', 'Address']

df = df.drop(columns=list_var_toDelete)
```

Ungelabeldete Daten werden gelöscht

```python
unlabeled = df[df["Sale_price"].isna()].copy()
print(f"Der Datensatz hat {len(unlabeled)} ungelabeldete Einträge")
labeled = df[df["Sale_price"].notna()]
print(f"Der Datensatz hat {len(labeled)} gelabeldete Einträge")

# NaN-Werte der Spalze Sale_price werden gelöscht, da es sich hierbei um die Zielvariable hadnelt
df = df.dropna(subset=['Sale_price'])
```

Zusammenführung der Spalten Nbhd und nbhd

```python
check_nbhd = df["Nbhd"].notna() & df["nbhd"].notna() 
print(df[check_nbhd])

df["nbhd"] = df["nbhd"].fillna(df["Nbhd"])
df = df.drop(columns=['Nbhd'])
print(df["nbhd"].isna().sum())
```

Zusammenführung der Spalten Nr_of_rms und Rooms

```python
check_room = df["Nr_of_rms"].notna() & df["Rooms"].notna() 
print(df[check_room])


df["Nr_of_rms"] = df["Nr_of_rms"].fillna(df["Rooms"])
df = df.drop(columns=['Rooms'])

print(df["Nr_of_rms"].isna().sum())
```

Zusammenführung der Spalten Taxkey und taxkey

```python
check_nbhd = df["Taxkey"].notna() & df["taxkey"].notna() 
print(df[check_nbhd])

df["taxkey"] = df["taxkey"].fillna(df["Taxkey"])
df = df.drop(columns=['Taxkey'])

print(df["taxkey"].isna().sum())
```

Zusammenführung der Spalten FinishedSqft und Fin_sqft

```python
# Anpassung der Punktation inkl. Änderung des Datentyp von Lotsize in Int
df['FinishedSqft'] = pd.to_numeric(df['FinishedSqft'].astype(str).str.replace(',', '', regex=False),errors='coerce').astype('float')

check_nbhd = df["Fin_sqft"].notna() & df["FinishedSqft"].notna() 
print(df[check_nbhd])

df["FinishedSqft"] = df["FinishedSqft"].fillna(df["Fin_sqft"])
df = df.drop(columns=['Fin_sqft'])

print(df["FinishedSqft"].isna().sum())
```

Umwandlung der Spalten mit eig. gannzahligen Datenformat

```python
 # Spalten die eig. ganzzahlig sind, werden auf das Datenformat Int geändert
for col in df.select_dtypes(include="float"):
    if (df[col].dropna() % 1 == 0).all():  # Prüfung, ob alle Werte auf .0 enden (ganzzahlig sind), dabei werden NaN Werte nicht betrachtet
        df[col] = df[col].astype("Int64")
```

Anpassung der Spalte Lotsize (Punktation, Datenformat, Extremwerte)

```python
# Anpassung der Punktation inkl. Änderung des Datentyp von Lotsize in Int
df['Lotsize'] = pd.to_numeric(df['Lotsize'].astype(str).str.replace(',', '', regex=False),errors='coerce').astype('Int64')

# Extremwerte der Spalte Lotsize werden gelöscht
df.loc[df['Lotsize'] > 200000, 'Lotsize'] = np.nan
df = df.dropna(subset=['Lotsize'])
```

Anpassung der Spalte Sale_price (Punktation, Datenformat, Extremwerte)

```python
# Anpassung der Punktation inkl. Änderung des Datentyp von Sale_price in Int
df['Sale_price'] = pd.to_numeric(df['Sale_price'].astype(str).str.replace('[^\d]', '', regex=True), errors='coerce').astype('Int64')

# Extremwerte der Spalte Sale_price werden gelöscht
df.loc[df['Sale_price'] >= 5000000, 'Sale_price'] = np.nan
df = df.dropna(subset=['Sale_price'])
```

Anpassung der Spalte Extwall (Nan-Werte, Datenformat)

```python
df['Extwall'] = df['Extwall'].fillna('Unknown')

# Anpssung der Satzzeichen von extwall
df["Extwall"] = df["Extwall"].str.replace("-", "_").str.replace("/", "_").str.replace(" ", "")
```

Anpassung der Spalte Units (Extremwerte)

```python
# Extremwerte der Spalte units werden gelöscht
df.loc[df['Units'] > 12, 'Units'] = np.nan
df = df.dropna(subset=['Units'])
```

Anpassung der Spalte Hbath (Extremwerte)

```python
# Extremwerte der Spalte hbath werden gelöscht
df.loc[df['Hbath'] > 9, 'Hbath'] = np.nan
df = df.dropna(subset=['Hbath'])
```

Anpassung der Spalte Year_Built (unrealistische Werte)

```python
# Unrealistische Werte auf NaN setzten
df.loc[df['Year_Built'] < 1000, 'Year_Built'] = np.nan
```

Anpassung der Spalte Bdrms (unrealistische Werte)

```python
df.loc[df['Bdrms'] > 100, 'Bdrms'] = np.nan
```
