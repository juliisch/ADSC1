## Datenaufbereitung

```python
# Duplikate löschen
df = df.drop_duplicates()
```

In der vorherigen Dateneinsicht haben sich folgende Aufbereitsungsschritte für die einzelnen Variablen ergeben:

- PropertyID: Kann gelöscht werden, da diese Variable keine sinnvollen Informationen beinhaltet.
- PropType: Fehlende Einträge (NaN) werden auf "Unknown" gesetzt.
- taxkey: Wird gelöscht, da der Steuerschlüssel sich 
- Address: Fehlende Einträge (NaN) werden auf "Unknown" gesetzt.
- CondoProject: Fehlende Einträge (NaN) werden auf "None" gesetzt
- District
- nbhd
- Style: Fehlende Einträge (NaN) werden auf "Unknown" gesetzt.
- Extwall: Fehlende Einträge (NaN) werden auf "Unknown" gesetzt.
- Stories
- Year_Built: Einträge bei denen eine extreme niedrige Jahresanzahl (n < 1000) auftritt werden gelöscht.
- Rooms
- FinishedSqft
- Units
- Bdrms: Einträge bei denen eine extreme und unnatürliche Anzahl an Schlafzimmern (n > 100) auftrit werden gelöscht.
- Fbath
- Hbath
- Lotsize: Datentyp in Integer ändern.
- Sale_date
- Sale_price: Fehlende Einträge (NaN) werden gelöscht, da es sich bei dieser Variable um die Zielvariable handelt. Datentyp in Integer ändern.

CondoProject		84.185703

```python
df.loc[df['Year_Built'] < 1000, 'Year_Built'] = np.nan
df.loc[df['Bdrms'] > 100, 'Bdrms'] = np.nan


df['PropType'] = df['PropType'].fillna('Unknown')
df['Style'] = df['Style'].fillna('Unknown')
df['Extwall'] = df['Extwall'].fillna('Unknown')
df['Address'] = df['Address'].fillna('Unknown')
df['CondoProject'] = df['CondoProject'].fillna('None')

# NaN-Werte der Spalze Sale_price werden gelöscht, da es sich hierbei um die Zielvariable hadnelt
df = df.dropna(subset=['Sale_price'])
```

```python
df.info()
```

```python
df = df.drop(columns=['PropertyID', 'Taxkey','taxkey', 'nbhd', 'Rooms', 'FinishedSqft', 'Sale_date'])
```

```python
 # Spalten die eig. ganzzahlig sind, werden auf das Datenformat Int geändert
for col in df.select_dtypes(include="float"):
    if (df[col].dropna() % 1 == 0).all():  # Prüfung, ob alle Werte auf .0 enden (ganzzahlig sind), dabei werden NaN Werte nicht betrachtet
        df[col] = df[col].astype("Int64")

# Datentyp von Lotsize in Int ändern
df['Lotsize'] = pd.to_numeric(df['Lotsize'].astype(str).str.replace(',', '', regex=False),errors='coerce').astype('Int64')
# Datentyp von Sale_price in Int ändern
df['Sale_price'] = pd.to_numeric(df['Sale_price'].astype(str).str.replace('[^\d]', '', regex=True), errors='coerce').astype('Int64')

# Datentyp von Sale_date auf Datetime ändern
#df["Sale_date"] = pd.to_datetime(df["Sale_date"], format="%Y-%m", errors='coerce').dt.to_period("M")
```
